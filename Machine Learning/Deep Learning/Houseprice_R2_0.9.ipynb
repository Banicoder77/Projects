```python
import kagglehub
import pandas as pd
import torch.nn as nn
import torch
from torch.utils.data import Dataset,DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import matplotlib.pyplot as plt
%matplotlib qt5
```

```python
import os

path = kagglehub.dataset_download("mohamedafsal007/house-price-dataset-of-india")
print("Dataset downloaded to:", path)
data_file = os.path.join(path, "House Price India.csv")
df=pd.read_csv(data_file)
df
```

```python
df=df.drop(columns=['Date','id'])
df
```

```python
scaler = StandardScaler()
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns=['Price']),df['Price'],test_size=0.2,random_state=42)
x=torch.tensor(scaler.fit_transform(x_train.values),dtype=torch.float32)
y=torch.tensor(y_train.values,dtype=torch.float32).view(-1,1)
print(x.shape,y.shape)
```

```python
import torch
import torch.nn as nn

class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()

        self.fc1 = nn.Linear(20, 512)
        self.fc2 = nn.Linear(512, 512)
        self.fc3 = nn.Linear(512, 256)
        self.fc4 = nn.Linear(256, 256)
        self.fc5 = nn.Linear(256, 128)
        self.fc6 = nn.Linear(128, 128)
        self.fc7 = nn.Linear(128, 64)
        self.fc8 = nn.Linear(64, 32)
        self.fc9 = nn.Linear(32, 1) 
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.2)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.dropout(x)

        x = self.relu(self.fc2(x))
        x = self.dropout(x)

        x = self.relu(self.fc3(x))

        x = self.relu(self.fc4(x))
        x = self.dropout(x)

        x = self.relu(self.fc5(x))

        x = self.relu(self.fc6(x))
        x = self.dropout(x)

        x = self.relu(self.fc7(x))
        x = self.relu(self.fc8(x))

        x = self.fc9(x)
        return x
loss_fn=nn.MSELoss()
```

```python
model=Model()
```

```python
optim=torch.optim.Adam(model.parameters(),lr=0.001)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optim,
    mode='min',
    factor=0.5,     # reduce LR by half
    patience=10,     
)
```

```python
class mDataset(Dataset):
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def __len__(self):
        return len(self.x)

    def __getitem__(self, idx):
        return self.x[idx],self.y[idx]
```

```python
data=mDataset(x,y)
loader=DataLoader(data,batch_size=32,shuffle=True)
```

```python
fig,axs=plt.subplots(1,1)
axs.set_title('Combined Loss Plot')
loss1=[]
loss2=[]
axs.set_xlabel('Epochs')
axs.set_ylabel('Loss')
l1=axs.plot(loss1,label='Train Loss')
l2=axs.plot(loss2,label='Test Loss')
txt=axs.text(0.5,0.9,'',transform=axs.transAxes,ha='center')
axs.legend()
plt.show()
```

```python
x_test=torch.tensor(scaler.transform(x_test.values),dtype=torch.float32)
y_test=torch.tensor(y_test.values,dtype=torch.float32).view(-1,1)
```

```python
def evaluate(model,x,y):
    model.eval()
    with torch.no_grad():
        y_pred=model(x)
        loss=loss_fn(y_pred,y)
        return loss.item()
```

```python
c=0

while True:
    model.train()
    c+=1
    loss=0
    for i,j in tqdm(loader,total=len(loader),desc=f'Epoch {c}'):
        y_pred=model(i)
        lloss=loss_fn(y_pred,j)
        optim.zero_grad()
        lloss.backward()
        loss+=lloss.item()
        optim.step()
    print(f'Epoch {c} Loss:{loss/len(loader)}')
    gg=evaluate(model,x_test,y_test)
    loss1.append(evaluate(model,x,y))
    loss2.append(gg)
    scheduler.step(gg)
    current_lr = scheduler._last_lr[0] if hasattr(scheduler, "_last_lr") else optimizer.param_groups[0]['lr']
    print(f"Current LR: {current_lr:.6f}")

    l1[0].set_ydata(loss1)
    l1[0].set_xdata(range(1,len(loss1)+1))
    l2[0].set_ydata(loss2)
    l2[0].set_xdata(range(1,len(loss2)+1))
    txt.set_text(f'Best Loss: {min(loss2):.4f} at Epoch {loss2.index(min(loss2))+1}')
    plt.pause(0.001)
    axs.relim()
    axs.autoscale_view()
    
```

```python

model.eval()
with torch.no_grad():
    y_pred=model(x_test)
    test_loss=loss_fn(y_pred,y_test)
    print("Test Loss:",test_loss.item())
    for i in range(len(x_test)):
        print(f'Predicted:{y_pred[i].item():.2f}, Actual:{y_test[i].item():.2f}')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
# Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)

# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# Root Mean Squared Error
rmse = np.sqrt(mse)

# RÂ² score
r2 = r2_score(y_test, y_pred)
print(f'Mean Absolute Error: {mae:.2f}')
print(f'Mean Squared Error: {mse:.2f}')
print(f'Root Mean Squared Error: {rmse:.2f}')
print(f'RÂ² Score: {r2:.2f}')
```
