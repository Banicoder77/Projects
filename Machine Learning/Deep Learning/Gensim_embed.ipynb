```python
from gensim.models import KeyedVectors
import spacy

nlp = spacy.load("en_core_web_md")

model = KeyedVectors.load_word2vec_format(
    "/home/user456/Downloads/wiki_giga_2024_100_MFT20_vectors_seed_2024_alpha_0.75_eta_0.05.050_combined.txt",binary=False,no_header=True
)
```

```python
def token(text):
    li=[]
    do=nlp(text)
    for x in do.sents:
        tokens=[t.lemma_.lower() 
                for t in x
                ]
        if tokens:
            li.append(tokens)
    return li
def similar_finder(word):
    return model.most_similar(word,topn=10)
def vector(word):
    return model[word]
    
z=token("I scored 98 marks in 2025 exam.")
print(z)
a=vector("girl")-vector("boy")+vector("king")
print(model.similar_by_vector(a,topn=10))
```
